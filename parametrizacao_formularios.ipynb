# Parametriza√ß√£o de Formul√°rios

## Importar biblioteca

```python
import json
```

## Ler formulario

```python
print("Lendo formul√°rio base...")

with open("formulario_base.json", "r", encoding="utf-8") as arquivo:
    dados = json.load(arquivo)

formulario = dados["formulario"]

print("Campos existentes:", len(formulario))
formulario
```

## Gerar novo ID

```python
ids_existentes = [campo["id"] for campo in formulario]
novo_id = max(ids_existentes) + 1

print("IDs existentes:", ids_existentes)
print("Novo ID:", novo_id)
```

## Criar novo campo

```python
novo_campo = {
    "id": novo_id,
    "nome": "Telefone",
    "tipo": "texto",
    "obrigatorio": False
}

novo_campo
```

## Adicionar campo ao formul√°rio

```python
formulario.append(novo_campo)

print("Campo adicionado com sucesso.")
print("Total de campos agora:", len(formulario))
formulario
```

## Salvar formul√°rio final

```python
dados_atualizados = {
    "formulario": formulario
}

with open("formulario_final.json", "w", encoding="utf-8") as arquivo:
    json.dump(dados_atualizados, arquivo, indent=4, ensure_ascii=False)

print("Formulario_final.json criado.")
```

## Valida√ß√£o

```python
print("‚úÖ Validando arquivo gerado: formulario_final.json")

with open("formulario_final.json", "r", encoding="utf-8") as arquivo:
    dados_final = json.load(arquivo)

# 1) Confere se a chave principal existe
if "formulario" not in dados_final:
    raise KeyError("Arquivo final n√£o cont√©m a chave 'formulario'.")

formulario_final = dados_final["formulario"]

# 2) Confere quantidade de campos
print("Campos no arquivo final:", len(formulario_final))

# 3) Confere se o campo 'Telefone' existe
campo_telefone = next((c for c in formulario_final if c.get("nome") == "Telefone"), None)

if campo_telefone is None:
    raise ValueError("Campo 'Telefone' n√£o foi encontrado no arquivo final.")

print("‚úÖ Campo 'Telefone' encontrado:", campo_telefone)

# 4) Confere IDs √∫nicos
ids = [c["id"] for c in formulario_final]
if len(ids) != len(set(ids)):
    raise ValueError("H√° IDs repetidos no arquivo final.")

print("‚úÖ IDs √∫nicos confirmados:", ids)
print("üéâ Valida√ß√£o conclu√≠da com sucesso! Projeto finalizado.")
```
